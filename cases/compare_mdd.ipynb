```python
from matplotlib import pyplot as plt
import numpy as np
from strava import activity
```

```python
%pylab inline
```

```python
# Set the activities id, can be found in the link to activity e.g. https://www.strava.com/activities/628508858

id_mdd28 = '163301368'
id_mdd30 = '628508858'

# Create activity object, if athlete is not specified, it will be created based on ~/config/athlete.config file

mdd28 = activity.Activity(activity_id=id_mdd28)
mdd30 = activity.Activity(activity_id=id_mdd30)
```

```python
# Visualize Strava-like activity analysis 

mdd28.plot_strava_analysis_simple(title='mdd28')
```

```python
mdd30.plot_strava_analysis_simple(title='mdd30')
```

```python
# Create object that compares two activities

mdd_compare = activity.CompareActivities(mdd30, mdd28)
```

```python
# Visialize power overlap between two activities in the index space (more types of X axis will be added)

mdd_compare.plot_compare_power_streams(labels=['mdd30', 'mdd28'])
```

```python
# Analyze the power difference during the activity

mdd_compare.synchronize_activities_by_power_shifting()
```
